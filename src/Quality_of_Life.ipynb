```python
import pandas as pd
```

#### General

```python
df = pd.read_excel('World Dataviz Prize 2023 - Challenge Concepts & Datasets.xlsx',
                   sheet_name="What Just Happened Last 10 year", index_col=0)

df = df.drop(["Unnamed: 16"], axis=1)

df
```

```python
startQOL = 46
endQOL = 49

dfQOL = df.rename(columns={"index": "label", 2009: "2009", 2010: "2010", 
                              2011: "2011", 2012: "2012", 2013: "2013", 
                              2014: "2014", 2015: "2015", 2016: "2016",
                              2017: "2017", 2018: "2018", 2019: "2019", 
                              2020: "2020", 2021: "2021"})

dfQOL = dfQOL.fillna("")

dfHappy = dfQOL.iloc[df.index.get_loc("Happy Planet Index"), :].to_frame().T
dfQOL = dfQOL.iloc[startQOL:endQOL]
dfQOL = pd.concat([dfQOL, dfHappy])

dfQOL
```

```python
dfQOLTimeLine = dfQOL[["2009",	"2010",	"2011",	"2012",	"2013",	"2014",	"2015",
                       "2016",	"2017", "2018",	"2019",	"2020",	"2021"]].T

dfQOLTimeLine
```

```python
dfQOLTimeLine.to_csv('Last 10 year - QoL Timeline.csv', index=True)
```

```python
dfQOL = dfQOL[["data",	"metric", "10 year change",	"source name",	"source link"]]
dfQOL
```

```python
dfQOL.to_csv('Last 10 year - QoL.csv', index=True)
```
