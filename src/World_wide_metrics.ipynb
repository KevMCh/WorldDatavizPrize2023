```python
import re
import pandas as pd
```

#### General

```python
df = pd.read_excel('World Dataviz Prize 2023 - Challenge Concepts & Datasets.xlsx',
                   sheet_name="Dashboard of the Present Future", index_col=0)

df
```

```python
dfCountries = df[["ISO Country code", "population", "surface area (Km2)",
                  "% of population in extreme poverty",
                  "% of population with access to electricity",
                  "happy planet index", "happy planet index.1",
                  "human development index",	"human development index.1",
                  "unemployment (%)", "Unnamed: 49", "Economic Growth", 
                  "Unnamed: 23"]]

dfCountries
```

#### Countries

```python
dfCountries = dfCountries.rename(columns={"population": "Population" + " (" + 
                            str(dfCountries.loc['data year', 'population']) + ")",
                            "surface area (Km2)": "Surface area (Km2)" + " (" + 
                            str(dfCountries.loc['data year', 'surface area (Km2)'])+ ")",
                            "% of population in extreme poverty": "% of population in extreme poverty" + " (" + 
                            str(dfCountries.loc['data year', '% of population in extreme poverty']) + ")",
                            "% of population with access to electricity": "% of population with access to electricity" + " (" + 
                            str(dfCountries.loc['data year', '% of population with access to electricity']) + ")",
                            "happy planet index": "Happy planet index" + " (" +
                            str(dfCountries.loc['data year', 'happy planet index']) + ")",
                            "happy planet index.1": "Happy planet index" + " (" +
                            str(dfCountries.loc['data year', 'happy planet index.1']) + ")",
                            "human development index": "Human dev index" + " (" +
                            str(dfCountries.loc['data year', 'human development index']) + ")",
                            "human development index.1": "Human dev index" + " (" +
                            str(dfCountries.loc['data year', 'human development index.1']) + ")",
                            "unemployment (%)": "Unemployment (%)" + " (" +
                            str(dfCountries.loc['data year', 'unemployment (%)']) + ")",
                            "Unnamed: 49": "Unemployment (%)" + " (" +
                            str(dfCountries.loc['data year', 'Unnamed: 49']) + ")",
                            "Economic Growth": "Economic Growth" + " (" +
                            str(dfCountries.loc['data year', 'Economic Growth']) + ")",
                            "Unnamed: 23": "Economic Growth" + " (" +
                            str(dfCountries.loc['data year', 'Unnamed: 23']) + ")"
                          })

dfCountries = dfCountries.iloc[4:178]

dfCountries
```

```python
dfCountries.to_csv('Countries info.csv', index=True)
```

#### Comparative metrics

```python
dfCountriesComp = dfCountries[['Happy planet index (2016.0)', 'Happy planet index (2019)',
       'Human dev index (2017.0)', 'Human dev index (2021)',
       'Unemployment (%) (2018.0)', 'Unemployment (%) (2021.0)',
       'Economic Growth (2018.0)', 'Economic Growth (2021 or latest)']]

dfCountriesComp = dfCountriesComp.stack().reset_index(1)
dfCountriesComp.columns = ['Metric', 'Value']
dfCountriesComp
```

```python
dfCountriesComp['Year'] = dfCountriesComp['Metric'].str.extract("(\\d{4})")
dfCountriesComp['Metric'] = dfCountriesComp['Metric'].str.extract("(.*)\(\\d{4}")
dfCountriesComp
```

```python
dfCountriesComp.to_csv('Countries info comp.csv', index=True)
```
