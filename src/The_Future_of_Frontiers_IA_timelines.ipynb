```python
import pandas as pd
```

#### General

```python
df = pd.read_excel('World Dataviz Prize 2023 - Challenge Concepts & Datasets.xlsx',
                   sheet_name="The Future of Frontiers")

df
```

#### IA Section

```python
startAI = 34
endAI = 39

dfIA = df.iloc[startAI:endAI]
dfIA = dfIA.rename(columns={"now": "2021", 2025.0: "2025", 2030.0: "2030", 
                            2040.0: "2040", 2050.0: "2050"})

dfIA = dfIA.fillna("")

dfIA = dfIA[["Innovation / Development", "2021", "2025", "2030", "2040", "2050"]]

dfIA
```

```python
for i in range(len(dfIA)):
  dfTmp = dfIA[["2021", "2025", "2030", "2040", "2050"]]
  dfTmp = dfTmp.iloc[i].to_frame().reset_index()
  dfTmp = dfTmp.rename(columns={"index": "start", 34: "state", 35: "state", 
                                36: "state", 37: "state", 38: "state"})

  dfTmp["end"] = ["2025", "2030", "2040", "2050", "2060"]
  
  dfTmp["start"] = pd.to_numeric(dfTmp["start"])
  dfTmp["state"] = dfTmp["state"].astype(str)
  dfTmp["end"] = pd.to_numeric(dfTmp["end"])

  dfTmp.to_csv('The Future of Frontiers - ' + \
                       dfIA.iloc[i]['Innovation / Development'] + '.csv', index=False)
```
